```python
#Ingreso total de las tiendas

def numero_tienda(nombre_tienda):
    return list(tiendas.keys()).index(nombre_tienda) + 1

def suma_columna(nombre_tienda, columna):
    if nombre_tienda not in tiendas:
        print("La tienda no existe")
        return

    num_t = numero_tienda(nombre_tienda)
    ingreso_t = sum(tiendas[nombre_tienda][columna])

    print(f'El ingreso total de la tienda {num_t} ({nombre_tienda}) es: ${ingreso_t}')
```

# 2. Ventas por categoría

```python
#Cantidad de productos vendidos por categoría en cada tienda
def venta_categoria(nombre_tienda):
  if nombre_tienda not in tiendas:
        print("La tienda no existe")
        return

  num_t = numero_tienda(nombre_tienda)

  print(f'Las ventas por categoría de la tienda {num_t} ({nombre_tienda}), de mayor a menor son:')
  return tiendas[nombre_tienda].groupby('Categoría del Producto')[['Precio']].sum().sort_values(by='Precio',ascending=False)
```

```python
def format_millions(value):
    return f'{value/1e6:.1f} M'
```

```python
import matplotlib.pyplot as plt
import numpy as np

# -----------------------------
# Datos
# -----------------------------
category_names = [
    'Electrónicos', 'Electrodomésticos', 'Muebles',
    'Instrumentos musicales', 'Deportes y diversión',
    'Juguetes', 'Artículos para el hogar', 'Libros'
]

results = {
    'Tienda 1': [429493500.0, 363685200.0, 187633700.0, 91299000.0, 39290000.0, 17995700.0, 12698400.0, 8784900.0],
    'Tienda 2': [410831100.0, 348567800.0, 176426300.0, 104990300.0, 34744500.0, 15945400.0, 14746900.0, 10091200.0],
    'Tienda 3': [410775800.0, 329237900.0, 201072100.0, 77380900.0, 35593100.0, 19401100.0, 15060000.0, 9498700.0],
    'Tienda 4': [409476100.0, 283260200.0, 192528900.0, 75102400.0, 33350100.0, 20262200.0, 15074500.0, 9321300.0]
}

# -----------------------------
# Formateo a millones
# -----------------------------
def format_millions(value):
    return f'{value/1e6:.1f} M'

# -----------------------------
# Función de gráfico
# -----------------------------
def survey(results, category_names):

    labels = list(results.keys())
    data = np.array(list(results.values()))
    data_cum = data.cumsum(axis=1)

    # 🎨 Paleta con muchos colores
    category_colors = plt.colormaps['tab20'](
        np.linspace(0, 1, data.shape[1])
    )

    # 📐 Figura ampliada
    fig, ax = plt.subplots(figsize=(18, 9))

    ax.invert_yaxis()
    ax.xaxis.set_visible(False)
    ax.set_xlim(0, np.sum(data, axis=1).max())

    # -----------------------------
    # Barras y etiquetas
    # -----------------------------
    for i, (colname, color) in enumerate(zip(category_names, category_colors)):
        widths = data[:, i]
        starts = data_cum[:, i] - widths

        rects = ax.barh(
            labels,
            widths,
            left=starts,
            height=0.65,
            label=colname,
            color=color
        )

        # Etiquetas con línea cuando no caben
        for rect, value in zip(rects, widths):
          y = rect.get_y() + rect.get_height() / 2
          x_center = rect.get_x() + rect.get_width() / 2

          # Umbral: barra pequeña
          if rect.get_width() < 0.05 * ax.get_xlim()[1]:
              ax.text(
                  x_center,
                  y,
                  format_millions(value),
                  ha='center',
                  va='center',
                  fontsize=8,
                  color='white',
                  rotation=90
              )
          else:
              ax.text(
                  x_center,
                  y,
                  format_millions(value),
                  ha='center',
                  va='center',
                  fontsize=9,
                  color='white'
              )

    # -----------------------------
    # Leyenda
    # -----------------------------
    ax.legend(
        ncols=len(category_names),
        loc='upper center',
        bbox_to_anchor=(0.5, 1.12),
        fontsize=10,
        frameon=True
    )

    ax.set_title('Distribución de ventas por categoría y tienda', fontsize=15)

    plt.tight_layout(rect=[0, 0, 1, 0.88])
    plt.show()

    return fig, ax


# -----------------------------
# Ejecutar
# -----------------------------
survey(results, category_names)
```

# 3. Calificación promedio de la tienda


```python
#Satisfacción del cliente con los productos vendidos

def media_columna(nombre_tienda,columna):
  if nombre_tienda not in tiendas:
        print("La tienda no existe")
        return
  num_t = numero_tienda(nombre_tienda)

  calif_tienda = tiendas[nombre_tienda][columna].mean()
  print(f'El promedio de {columna} de la tienda {num_t} ({nombre_tienda}) es: {round(calif_tienda,2)}')
```

# 4. Productos más y menos vendidos

```python
#Función de contabilización

def conteo_columna(nombre_tienda,columna):
  if nombre_tienda not in tiendas:
        print("La tienda no existe")
        return

  conteo = tiendas[nombre_tienda][columna].value_counts().sort_values(ascending=False)
  mas_frecuente = conteo.index[0]
  menos_frecuente = conteo.index[-1]

  print(f'{columna} más frecuente fue {mas_frecuente} con {conteo.iloc[0]} veces')
  print(f'{columna} menos frecuente fue {menos_frecuente} con {conteo.iloc[-1]} veces')

  return conteo.plot(kind='barh',figsize=(12,8),color='red')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
ventas_totales = [1150880400.0, 1116343500.0, 1098019600.0, 1038375700.0]
envios = [26018.61, 25216.24, 24805.68, 23459.46]
calificaciones = [3.98, 4.04, 4.05, 4.0]

etiquetas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

ventas_n = np.array(ventas_totales) / max(ventas_totales)
envios_n = np.array(envios) / max(envios)
calif_n = np.array(calificaciones) / max(calificaciones)

#Figura
plt.figure(figsize=(12, 7))

plt.stackplot(
    etiquetas,
    ventas_n,
    envios_n,
    calif_n,
    labels=['Ventas (norm)', 'Envíos (norm)', 'Calificaciones (norm)']
)

# Etiquetas y título
plt.title('Stackplot de ventas, envíos y calificaciones normalizados por tienda', fontsize=14)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Valores', fontsize=12)

plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

# Informe final: Recomendación de tienda a vender – Cadena Alura Store

## Introducción

El presente informe tiene como objetivo apoyar al **Sr. Juan**, propietario de la cadena **Alura Store**, en la toma de decisión sobre **qué tienda vender** con el fin de obtener capital para iniciar un nuevo emprendimiento.

Para ello, se realizó un análisis integral del desempeño de las cuatro tiendas que conforman la cadena, considerando múltiples factores clave: **ingresos totales**, **categorías de productos más y menos vendidas**, **productos con mayor y menor frecuencia de venta**, **calificaciones promedio de los clientes** y **costes de envío promedio**.  
La evaluación se basa en datos cuantitativos y en las visualizaciones generadas durante el análisis.

---

## Desarrollo del análisis

### 1. Ingresos totales por tienda

El análisis de los ingresos totales muestra que todas las tiendas generan montos elevados; sin embargo, existen diferencias relevantes:

- **Tienda 1** presenta el **mayor ingreso total** de la cadena.
- **Tienda 2** y **Tienda 3** mantienen ingresos competitivos, aunque ligeramente por debajo.
- **Tienda 4** registra el **menor ingreso total**, lo que indica un menor aporte económico global.

**Conclusión parcial:** Desde la perspectiva de ingresos, Tienda 4 es la menos rentable dentro de la cadena.

---

### 2. Categorías de productos más y menos vendidas

A partir de los gráficos de barras apiladas se observa que:

- Las categorías **Electrónicos** y **Electrodomésticos** concentran la mayor proporción de ventas en todas las tiendas.
- Las categorías con menor participación son **Libros**, **Artículos para el hogar** y **Juguetes**.
- **Tienda 1** muestra una distribución equilibrada y sólida en las categorías de mayor demanda.
- **Tienda 4**, aunque mantiene presencia en categorías principales, presenta una menor contribución total en casi todas ellas.

**Conclusión parcial:** Tienda 1 y Tienda 2 presentan un mejor posicionamiento por categorías; Tienda 4 muestra menor fortaleza estructural.

---

### 3. Productos más y menos vendidos por tienda

El análisis de frecuencia de productos revela diferencias importantes en el comportamiento de ventas:

- **Tienda 1**
  - Producto más frecuente: *Microondas* (60 ventas)
  - Producto menos frecuente: *Celular ABXY* (33 ventas)

- **Tienda 2**
  - Producto más frecuente: *Iniciando en programación* (65 ventas)
  - Producto menos frecuente: *Juego de mesa* (32 ventas)

- **Tienda 3**
  - Producto más frecuente: *Kit de bancas* (57 ventas)
  - Producto menos frecuente: *Bloques de construcción* (35 ventas)

- **Tienda 4**
  - Producto más frecuente: *Cama box* (62 ventas)
  - Producto menos frecuente: *Guitarra eléctrica* (33 ventas)

Si bien Tienda 4 cuenta con un producto muy frecuente, su **diversidad y volumen total de ventas** es inferior al de las demás tiendas.

**Conclusión parcial:** La fortaleza de Tienda 4 depende de pocos productos, lo que representa un mayor riesgo comercial.

---

### 4. Calificaciones promedio de los clientes

Las calificaciones promedio de los clientes muestran que:

- Todas las tiendas mantienen evaluaciones cercanas a **4.0**, reflejando una satisfacción general positiva.
- **Tienda 3** presenta la calificación promedio más alta.
- **Tienda 4** no destaca particularmente en este aspecto frente a las demás.

**Conclusión parcial:** La satisfacción del cliente no representa una ventaja competitiva clara para Tienda 4.

---

### 5. Coste de envío promedio

El análisis del coste de envío promedio indica que:

- **Tienda 4** presenta el coste de envío más bajo.
- **Tienda 1** y **Tienda 2** tienen costes ligeramente mayores, pero acordes con su volumen de ventas.
- La diferencia en costes de envío no compensa la menor generación de ingresos de Tienda 4.

**Conclusión parcial:** Un menor coste de envío no es suficiente para justificar la permanencia de una tienda con menor rentabilidad.

---

## Conclusión y recomendación final

Tras analizar de forma integral todos los factores evaluados —ingresos totales, desempeño por categorías, productos más y menos vendidos, calificaciones de clientes y costes de envío— se concluye que:

### **La tienda que el Sr. Juan debería vender para iniciar su nuevo emprendimiento es la _Tienda 4_.**

### Justificación

- Es la tienda con **menores ingresos totales** dentro de la cadena.
- Presenta una **menor fortaleza global por categorías** de productos.
- Su desempeño depende de **pocos productos altamente vendidos**, lo que implica mayor vulnerabilidad.
- No destaca significativamente en **calificaciones de clientes**.
- Aunque tiene el **menor coste de envío**, este beneficio no compensa su bajo aporte económico total.

En conjunto, **vender la Tienda 4 permitiría al Sr. Juan liberar capital sin comprometer significativamente la estabilidad y rentabilidad general de la cadena Alura Store**, conservando las tiendas con mejor desempeño para el crecimiento futuro.
